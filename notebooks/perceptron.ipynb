```python
import numpy as np
```

```python
W1 = np.array([
    [0.2, -0.3],
    [0.4, 0.1],
])
x = np.array([0, 1])
b1 = np.array([-0.4, 0.2])
```

```python
y1 = np.matmul(W1, x[:, None]) + b1[:, None]
y1
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
y2 = sigmoid(y1)
y2
```

```python
W2 = np.array([[-0.3, -0.2]])
b2 = np.array([0.1])
```

```python
y3 = np.matmul(W2, y2) + b2[:, None]
y3
```

```python
y4 = sigmoid(y3)
y4
```

```python
def mse(input, target):
    return np.square(input - target).mean()
```

```python
loss = mse(y4, 1)
loss
```

```python
def derivative_mse(input, target):
    return (2 * (input - target)).mean()
```

```python
delta_y = derivative_mse(y4, 1)
delta_y
```

```python
def derivate_sigmoid(x):
    return x * (1 - x)
```

```python
delta_y3 = delta_y * derivate_sigmoid(y4)
delta_y3
```

```python
delta_w2 = delta_y3 * (np.matmul(delta_y3, y2.transpose()))

delta_w2
```

```python
delta_b2 = delta_y3 * delta_y3
delta_b2
```

```python
delta_y2 = delta_y3 * np.matmul(W2.transpose(), delta_y3)
delta_y2
```

```python
delta_y1 = delta_y2 * derivate_sigmoid(y2)
delta_y1
```

```python
delta_w1 = delta_y1 * (np.matmul(delta_y1, y1.transpose()))

delta_w1
```

```python
delta_b1 = delta_y1 * delta_y1
delta_b1
```
