```python
import torch
import math
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
func = nn.Sigmoid()

x = np.arange(-6, 6, 0.01)
x_tensor = [torch.tensor([x_], requires_grad=True) for x_ in x]
y = [func(x_) for x_ in x_tensor]

z = []
for i in range(len(y)):
    y[i].backward()
    z.append(x_tensor[i].grad)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

fig.suptitle('Sigmoid')
plt.plot(x, y, label='Sigmoid')
plt.plot(x, z, label='Derivative of Sigmoid')
plt.legend()
plt.show()
```

```python
class Swish(nn.Module):
    def __init__(self):
        super(Swish, self).__init__()

    def forward(self, x):
        return x * F.sigmoid(x)

func = Swish()

x = np.arange(-6, 6, 0.01)
x_tensor = [torch.tensor([x_], requires_grad=True) for x_ in x]
y = [func(x_) for x_ in x_tensor]

z = []
for i in range(len(y)):
    y[i].backward()
    z.append(x_tensor[i].grad)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

fig.suptitle('Swish')
plt.plot(x, y, label='Swish')
plt.plot(x, z, label='Derivative of Swish')
plt.legend()
plt.show()
```

```python
class Softsign(nn.Module):
    def forward(self, x):
        return x / (1 + torch.abs(x))

func = Softsign()

x = np.arange(-6, 6, 0.01)
x_tensor = [torch.tensor([x_], requires_grad=True) for x_ in x]
y = [func(x_) for x_ in x_tensor]

z = []
for i in range(len(y)):
    y[i].backward()
    z.append(x_tensor[i].grad)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

fig.suptitle('Softsign')
plt.plot(x, y, label='Softsign')
plt.plot(x, z, label='Derivative of Softsign')
plt.legend()
plt.show()
```

```python
class Softplus(nn.Module):
    def forward(self, x):
        return torch.log(1 + torch.exp(x))

func = Softplus()

x = np.arange(-6, 6, 0.01)
x_tensor = [torch.tensor([x_], requires_grad=True) for x_ in x]
y = [func(x_) for x_ in x_tensor]

z = []
for i in range(len(y)):
    y[i].backward()
    z.append(x_tensor[i].grad)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

fig.suptitle('Softplus')
plt.plot(x, y, label='Softplus')
plt.plot(x, z, label='Derivative of Softplus')
plt.legend()
plt.show()
```
