```python
import wandb
import pandas as pd

api = wandb.Api()
```

```python
def dotter(mixed, key="", dots={}):
    if isinstance(mixed, dict):
        for (k, v) in mixed.items():
            dots = dotter(mixed[k], "%s.%s" % (key, k) if key else k, dots)
    else:
        dots[key] = mixed

    return dots
```

```python
runs_df = []

for run in api.runs("hiennguyen9874" + "/" + "rep-reid-v2"):
    run_df = dotter(
        {
            k: v
            for k, v in run.summary._json_dict.items()
            if not k.startswith("gradients")
        },
        dots=dict(),
    )

    run_df.update(
        dotter(
            {k: v for k, v in run.config.items() if not k.startswith("_")},
            dots=dict(),
        )
    )

    run_df.update({"name": run.name})

    runs_df.append(run_df)

for run in api.runs("hiennguyen9874" + "/" + "person-reid-v2"):
    run_df = dotter(
        {
            k: v
            for k, v in run.summary._json_dict.items()
            if not k.startswith("gradients")
        },
        dots=dict(),
    )

    run_df.update(
        dotter(
            {k: v for k, v in run.config.items() if not k.startswith("_")},
            dots=dict(),
        )
    )

    run_df.update({"name": run.name})

    runs_df.append(run_df)


runs_df = pd.DataFrame(runs_df)
```
